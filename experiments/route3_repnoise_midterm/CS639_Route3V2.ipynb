```python
%%bash
cd /content/cs639_project/Antidote

CUDA_VISIBLE_DEVICES=0 python train.py \
  --model_name_or_path meta-llama/Llama-2-7b-hf \
  --lora_folder ckpt/Llama-2-7b-hf_REDACTED_0.1_0.001 \
  --data_path PKU-Alignment/BeaverTails_dangerous \
  --fp16 True \
  --output_dir ckpt/sst2/Llama-2-7b-hf_REDACTED_0.1_0.001_f_0.2_5000_1e-4 \
  --num_train_epochs 20 \
  --per_device_train_batch_size 1 \
  --per_device_eval_batch_size 1 \
  --gradient_accumulation_steps 5 \
  --save_strategy steps \
  --save_steps 100000 \
  --save_total_limit 0 \
  --learning_rate 1e-4 \
  --weight_decay 0.1 \
  --warmup_ratio 0.1 \
  --lr_scheduler_type constant \
  --logging_steps 10 \
  --eval_steps 2000 \
  --cache_dir cache \
  --optimizer normal \
  --evaluation_strategy steps \
  --sample_num 5000 \
  --poison_ratio 0.2 \
  --label_smoothing_factor 0 \
  --benign_dataset data/sst2.json
```
